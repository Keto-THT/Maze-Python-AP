```python
import pygame
import time

# initialiser Pygame 
pygame.init()
pygame.font.init()


#créer une fenêtre de jeu 
##paramètres de la fenêtre
width = 640
height = 480
size = (width, height)

screen = pygame.display.set_mode(size)
pygame.display.set_caption('First game')
clock = pygame.time.Clock()


#gestion des événements : 
##fermeture de la fenêtre 
##pour le futur labyrinthe : gérer les flèches pour l'orientation dans le labyrinthe
##gérer les clics : imaginer des boutons de play, new game 


#gestion des événements : 
running = True 
while running : 
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    pygame.display.flip()

pygame.quit()


```
